```python
def sacarinfomcdonals3(xpath,col):
    url='https://mcdonalds.es/productos/sandwiches-principales'
    

    driver=webdriver.Chrome(PATH)
    driver.get(url)
    
    time.sleep(2)
    
    cookie=driver.find_element_by_xpath('//*[@id="onetrust-accept-btn-handler"]')
    cookie.click()  # Darle al boton de aceptar cookies
    
    time.sleep(3)
    


    producto=driver.find_element_by_xpath(f"{xpath}")
    producto.click()  #Buscar por xpath a la hamburguesa que darle
    
    time.sleep(2)
    
    nowurl= driver.current_url  #Cogeme el URL en el que estas actualmente
    
    html=req.get(nowurl).content
    
    soup=bs(html, 'html.parser')
    
    tabla=soup.find_all('table') #Busca dentro del tag tabla
    
    elementos = tabla[0].find_all('td') # Encuentrame cada tag td
    Bigmac=[f.text.strip().split('<td>') for f in elementos] #borra td del texto
    ahorasi= []
    for e in Bigmac:  # Por cada elemento en lista y por cada elemento dentro de estos 
        for u in e:
            if u == '':  # Si esta vacio borra
                del(u)
            elif u == 'Peso':   #Chapuzas de MCdonals
                del(u)
            elif u == '124Peso':
                del(u)
            elif u =='239g':
                del(u)
            elif u =='333g':
                del(u)
            elif u =='263g':
                del(u)
            elif u =='105g':
                del(u)
            elif u =='237,2g':
                del(u)
            elif u =='186,6g':
                del(u)
            elif u =='136g':
                del(u)
            elif u =='g':
                del(u)
            elif u =='137g':
                del(u)
            elif u =='127g':
                del(u)
            elif u =='218,4g':
                del(u)
            elif u =='170,4g':
                del(u)
            elif u =='264g':
                del(u)
            elif u =='248,2g':
                del(u)
            elif u =='239g':
                del(u)
            elif u =='279':
                del(u)
            elif u =='256g':
                del(u)
            elif u =='348g':
                del(u)
            elif u =='236,4g':
                del(u)
            elif u =='226g':
                del(u)
            else:
                ahorasi.append(u) # Si no metemelo aqui en esta lista
                
                
     #Creame este diccionario  ( Un poco chapuzalia)          
                
    fin = [ahorasi[6],ahorasi[10],ahorasi[14],ahorasi[18],ahorasi[21],ahorasi[25],ahorasi[29],ahorasi[33]]
    
    #Creame un dataframe con este diccionario y añademe la Columna 'Producto' con el valor que yo te diga
    #Creame una columna con el nombre del restaurante para poder unir en SQL
    
   
    fin.insert(0, f"{col}")
    fin.insert(9,'Mcdonals')
    
    
   
    
    
   
    return fin
```
